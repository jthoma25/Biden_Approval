```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
from matplotlib import pyplot as plt
import numpy as np
```

```python
#### APRIL 13 PIE CHART #####

#Data
Type_of_tweet = ['negative', 'positive', 'neutral']

number = [32, 34, 32]

colors = ("red", "blue", "grey")
```

```python
#Wedge width and color
wedge = {'linewidth':2, 'edgecolor':"black"}
```

```python
#percent function
def func(pct, allvalues):
    ab = int(pct/98.*np.sum(allvalues))
    return "{:.1f}%".format(pct, ab)
```

```python
#Plot
fig, ax = plt.subplots(figsize = (10, 7))
wedges, texts, autotexts = ax.pie(number, autopct = lambda pct: func(pct, number),
                                 labels = Type_of_tweet,
                                 colors = colors,
                                 shadow = True,
                                 startangle = 90,
                                 wedgeprops = wedge,
                                 textprops = dict(color = "black"))

ax.set_title("April 13th Pie Chart")
#plt.pie(number, labels = Type_of_tweet)
plt.show()
```
